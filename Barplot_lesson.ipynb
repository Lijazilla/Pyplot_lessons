```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ['India', 'Japan','Mexico', 'Colombia', 'Germany']
population = [1000,800,900,1000,300]
```

```python
plt.bar(country, population, width=0.5, color=['red','green','blue'])
plt.xticks(np.arange(5),('India', 'Japón', 'México', 'Colombia', 'Alemania'), rotation=45)
plt.title('Population by Country')
plt.xlabel('Country')
plt.ylabel('Population')
plt.show()
```

```python
plt.barh(country, population)
plt.show()
```
