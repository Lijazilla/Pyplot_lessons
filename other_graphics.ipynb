```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.randint(1,50,100)
```

## Histograma

```python
plt.hist(data, bins=10, width=3, histtype='bar')
plt.show()
```

```python
plt.hist(data, bins=10, width=3, histtype='step')
plt.show()
```

##BoxPlot

```python
data = np.append(data, 200)
plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=False)
plt.show()
```

##Scatter

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N)) **2
colors = np.random.rand(N)
```

```python
plt.scatter(x,y,s=area,c=colors,marker='o', alpha=0.5)
plt.show()
```
