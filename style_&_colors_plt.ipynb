```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
print(plt.style.available)
```

```python
plt.style.use('dark_background')
```

#Matlab

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x,x+1, 'r--')
ax.plot(x,x+2, 'b-')
ax.plot(x,x+3, 'y')
ax.plot(x,x+4, 'go:')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x,x+1,color='green', alpha=0.3, linewidth=5, linestyle='-')
ax.plot(x,x+2, color='blue', linewidth=3, linestyle='--')
ax.plot(x,x+3, color="#66ff66", linewidth=8, linestyle='dashed')
ax.plot(x,x+4, color="#ccffcc", linewidth=2, linestyle=':')
```
